# Act like a climate model 

## Example 1: Density profile 

```python
import numpy as np
heights = np.linspace(0,500, 11).T
density = [1015,1018, 1020, 1021, 1022, 1026, 1027, 1027.5, 1027.8, 1028, 1028.2]
old_density = [1015,1018, 1020, 1021, 1022, 1026, 1027, 1027.5, 1027.8, 1028, 1028.2]
```

```python
import matplotlib.pyplot as plt 
plt.style.use('seaborn')
plt.rcParams.update({'font.size': 20,
                     'xtick.labelsize' : 20,
                     'ytick.labelsize' : 20})
plt.figure(figsize=(8, 12))


plt.gca().invert_yaxis()
plt.plot(density,heights)
plt.ylabel('depth (m)',  fontsize= 25)
plt.xlabel('density kg m$^{-3}$',  fontsize= 25)
plt.savefig('ocean_profile.png')
plt.show()
```

```python

perturb = 1025

for idx,d in enumerate(density): 
    if perturb > d:
        pass
    else:
        density[idx] = perturb
        break
        
```

```python
plt.gca().invert_yaxis()
plt.plot(density,heights)
plt.plot(old_density,heights)
plt.show()
```

##  Example 3: Equator to pole gradient 

```python
lat= np.linspace(0,90,10)
temp= [30,27,25,20,8,0,-10,-15,-28,-30]

plt.style.use('seaborn')
plt.rcParams.update({'font.size': 20,
                     'xtick.labelsize' : 20,
                     'ytick.labelsize' : 20})
plt.figure(figsize=(12,7))
plt.plot(lat, temp)
plt.xlabel('lat ($\phi$)',  fontsize= 25)
plt.ylabel('temperature C$^\circ$', fontsize= 25)
plt.savefig('temp_eq_to_pole.png')
plt.show()
```
