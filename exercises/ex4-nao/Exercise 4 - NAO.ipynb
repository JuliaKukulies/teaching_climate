# North Atlantic Oscillation 

**What is the NAO?** 

https://www.youtube.com/watch?v=KOYJG7j4Iy8



**How is the NAO calculated ?**

*NAO Index* = SLP$_{Lisbon}$ - SLP$_{Stykkisholmur/Reykjavik}$

where SLP refers to the normalized sea level pressure of respective location. 

**Now, let's have a look at a timeseries of annual NAO indices** 

```python
# import python libraries 
import pandas as pd 
import numpy as np 


# read in a time series with NAO index 

nao= pd.read_csv('nao.csv', delimiter=r"\s+", header= None)

# do some stuff with the data, so it looks nice, e.g. column names in table
nao.index = nao.iloc[:,0].values
year = nao.index
nao = nao.iloc[:, 1::]
months= [ 'Jan','Feb', 'Mar',  'Apr', 'May', 'Jun', 'Jul','Aug', 'Sep', 'Oct', 'Nov', 'Dec', 'Annual']
nao.columns = months

# define NaN values 
nao[nao < -90 ] = np.nan
```

This is our our dataframe looks like 


```python
nao
```

```python
nao.shape
```

We have a data matrix which contains 197 rows and 13 columns. The rows represent the years and the columns represent the months (last column is the annual mean). 

```python
# python library for plotting 
import matplotlib.pyplot as plt
%matplotlib inline

# plot timeseries
plt.figure(figsize=(8,5))
annual_nao = nao.iloc[:, 12]
        
plt.bar(year, annual_nao , color=(annual_nao > 0).map({True: 'b', False: 'r'}))                                                                              
plt.xlim(np.nanmin(year), np.nanmax(year))
plt.ylabel('NAO index ')
plt.show()
```

## Can you identify years with strongly negative or positive NAO index values? 

```python
# get the year with the lowest value
year[annual_nao == np.nanmin(annual_nao)][0]


# can you find out what to type to get the max value?
```

```python
# look at monthly NAO index in year 2010 
nao2010= nao[nao.index == 2010]
nao2010
```

A quick look at the monthly NAO index for 2010 shows us that the index is especially strong during the winter months.
